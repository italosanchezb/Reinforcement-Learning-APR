## Variables

El elemento base dentro de Programación.

```python

x = 10

print(x)
print(type(x))
```

### Operaciones entre variables

```python
x = 10
y = 10

print(f"Suma: {x+y}")
print(f"Resta: {x-y}")
print(f"Multiplicación: {x*y}")
print(f"División: {x/y}")
print(f"Potenciación: {x**y}")
print(f"División Entera: {x//y}")
print(f"Módulo: {x%y}")

z = 2 + 1*3 - 10 / 2
print(f"Expresiones", z)
```

## Listas

Las listas se resumen en un conjunto indexado de elementos



```python
lista = [1,2,3,4,5]
print(lista)
```

### Metodos

Las listas, al ser un tipo de dato. Poseen "funciones" incorporadas.


```python
lista.append("Hola")
print(lista)

lista.remove(3)
print(lista)
```

### Operaciones compuestas con Listas (List Comprehensions)

```python
a = [1,2,3,4,5]
b = [x**2 for x in a]
print(b)

q = list(range(15))
q_impar = [x for x in q if x%2 != 0]
print(q)
print(q_impar)
```


## Funciones

Las funciones se conforman por bloques de código que pueden admitir argumentos. La función usual

$$
    f:X\to Y,
$$

donde $X$, y $Y$ dependerán de la necesidad de la función.

```python
def suma(a,b, **kwargs):
  return a+b

print(f"Función con Parámetros con retorno, {suma(1,2)}")

def f():
    print("Hola")

print(f"Función sin parámetros sin retorno ", {f()})
f()

def g():
    x = 10
    return x

print(f"Función sin parámetros con retorno ", {g()})

```

### Graficos (Matplotlib)

Dentro de Python, una de sus funciones fundamentales es la visualización de datos. Para ello, dentro del grupo de librerías que usaremos es Matplotlib.

Imagen1.png


```python
import matplotlib.pyplot as plt


x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

### Personalización

```python
fig, ax = plt.subplots(figsize = (8,6))
ax.plot(x,y)  # Grafico
ax.plot(x,2*y, color = "blue")  # Color
ax.plot(x,3 * y, linestyle = "--")  # Tipo de Linea
ax.set_xlabel("Eje X")  # Ejes
ax.set_ylabel("Eje Y")
ax.set_title("Grafica de Prueba")  # Titulo
plt.show()
```

### Multiple Graficos

```python
multi_fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (8,6), sharey = True)
ax[1,0].plot(x,y, color = "red", label = "G1")
ax[1,0].set_xlabel("Eje X")
ax[1,0].set_ylabel("Eje Y")
ax[1,0].set_title("Grafica de Prueba")
# Un comentario


x_0 = 100


ax[1,1].plot(x,2*y, color = "green", linestyle = "dotted", label = "G2")
ax[0,0].plot(x,3*y, color = "blue", linewidth = 0.1, label = "G3")
ax[0,1].plot(x,4*y, color = "black", linewidth = 2, label = "G4")
multi_fig.legend()
multi_fig.tight_layout()
plt.show()
```
